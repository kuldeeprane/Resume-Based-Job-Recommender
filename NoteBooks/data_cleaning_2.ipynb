```python
df = df.drop(columns=["Unnamed: 0"], errors="ignore")
```

```python
df['Title'] = df['Title'].astype(str).str.strip().str.lower()
df['Company'] = df['Company'].astype(str).str.strip().str.lower()
df['Location'] = df['Location'].astype(str).str.strip().str.title()
```

```python
df.head()  # Display the first few rows of the DataFrame
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['Experience'] = df['Experience'].fillna('0 Yrs')
```

```python
df['Short Skills'] = df['Short Skills'].fillna('None')
```

```python
df['Industry Type'] = df['Industry Type'].fillna('Unknown')
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])
df['Employment Type'] = df['Employment Type'].fillna(df['Employment Type'].mode()[0])
df['Role Category'] = df['Role Category'].fillna(df['Role Category'].mode()[0])
```

```python
# Step 1: Drop everything after the *last* occurrence of 'Key Skills'
df['Full JD'] = df['Full JD'].str.rpartition('Key Skills')[0].str.strip()
```

```python
print(df['Full JD'].iloc[11010])
```

```python
import re
# text cealing function

def clean_text(text):
    if pd.isna(text) or not isinstance(text, str):
        return ""

    # 1. Lowercase everything
    text = text.lower()
    
    # 2. Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    
    # 3. Remove unwanted characters (like #, *, etc. but keep useful punctuation if needed)
    text = re.sub(r'[^\w\s.,;:/()-]', '', text)
    
    # 4. Replace multiple spaces or line breaks with a single space
    text = re.sub(r'\s+', ' ', text)
    
    # 5. Remove leading/trailing whitespace
    text = text.strip()

    return text
```

```python
df['job_description_clean'] = df['Full JD'].apply(clean_text)
```
